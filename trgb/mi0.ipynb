```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# Selected optical TRGB calibrations
# -----------------------------------------------------------

data = [
    ("Cerny et al. 2020 (ω Cen, Gaia EDR3)",      -4.056, 0.100, "GC"),
    ("Soltis et al. 2021 (ω Cen, Gaia EDR3)",     -3.970, 0.060, "GC"),
    ("Jang et al. 2021 (NGC 4258)",               -4.043, 0.056, "Anchor"),
    ("Li et al. 2022 (Gaia EDR3)",                -3.910, 0.100, "Gaia"),
    ("Anderson et al. 2023 (LMC OSARG)",          -3.979, 0.030, "OSARG"),
    ("Hoyt et al. 2023 (LMC)",                    -4.038, 0.034, "Anchor"),
    ("Dixon et al. 2023 (Gaia DR3)",              -4.042, 0.051, "Gaia"),
    ("Koblischke et al. 2024 (SMC SARG)",         -4.024, np.hypot(0.041,0.029), "OSARG"),
    ("Koblischke et al. 2024 (SMC B-sequence)",   -4.057, np.hypot(0.019,0.029), "OSARG"),
    ("Udalski et al. 2025 (LMC)",                 -4.022, np.hypot(0.006,0.033), "Anchor"),
    ("Cruz Reyes et al. 2026 (Globular Clusters)",-3.948,0.036,"New"),
]

colors = {
    "Anchor":"tab:red",
    "Gaia":"tab:blue",
    "GC":"tab:orange",
    "OSARG":"tab:purple",
    "New":"tab:green"
}

fig, ax = plt.subplots(figsize=(9.5,6.5))

y = np.arange(len(data),0,-1)

for yy,(label,val,err,group) in zip(y,data):
    ax.errorbar(val,yy,
                xerr=err,
                fmt='o',
                color=colors[group],
                capsize=4,
                ms=6)

# labels
ax.set_yticks(y)
ax.set_yticklabels([d[0] for d in data],fontsize=9)

# guide lines
ax.axvline(-4.05,ls='--',lw=1,color='0.7')
ax.axvline(-4.00,ls='--',lw=1,color='0.7')

ax.set_xlim(-4.3,-3.7)
ax.set_ylim(0.5,len(data)+0.8)

ax.set_xlabel(r'$M_I(\mathrm{TRGB})$ [mag]',fontsize=12)
ax.set_title("Recent optical TRGB zero-point calibrations",fontsize=13)

# legend
from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0],[0],marker='o',color='w',markerfacecolor='tab:red',
           markersize=7,label='LMC / NGC4258'),
    Line2D([0],[0],marker='o',color='w',markerfacecolor='tab:blue',
           markersize=7,label='Gaia'),
    Line2D([0],[0],marker='o',color='w',markerfacecolor='tab:orange',
           markersize=7,label='Globular Clusters'),
    Line2D([0],[0],marker='o',color='w',markerfacecolor='tab:purple',
           markersize=7,label='OSARG'),
    Line2D([0],[0],marker='o',color='w',markerfacecolor='tab:green',
           markersize=7,label='New calibration'),
]

ax.legend(handles=legend_elements,
          loc='lower right',
          frameon=False,
          fontsize=9)

plt.tight_layout()

plt.savefig("TRGB_calibrations_forest.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

data = [
    # label, filter/telescope, M_TRGB, error, group
    ("Freedman et al. 2020", "HST F814W", -4.049, 0.047, "HST"),
    ("Jang et al. 2021",     "HST F814W", -4.043, 0.056, "HST"),
    ("Hoyt et al. 2023",     "HST F814W", -4.038, 0.034, "HST"),
    ("Dixon et al. 2023",    "HST F814W", -4.042, 0.051, "HST"),

    # przykład: wpisz swoje/adoptowane wartości JWST
    ("Freedman et al. 2024", "JWST F090W/F115W/F150W", -4.03, 0.05, "JWST"),
    ("Lee et al. 2024",      "JWST F115W/F150W",       -4.05, 0.06, "JWST"),
    ("Anand et al. 2024",    "JWST F115W/F150W",       -4.02, 0.05, "JWST"),
]

colors = {
    "HST": "tab:blue",
    "JWST": "tab:orange",
}

fig, ax = plt.subplots(figsize=(9.5, 5.5))

y = np.arange(len(data), 0, -1)

for yy, (label, filt, val, err, group) in zip(y, data):
    ax.errorbar(
        val, yy,
        xerr=err,
        fmt="o",
        color=colors[group],
        capsize=4,
        ms=6
    )

labels = [f"{label}   ({filt})" for label, filt, *_ in data]

ax.set_yticks(y)
ax.set_yticklabels(labels, fontsize=9)

ax.axvline(-4.05, ls="--", lw=1, color="0.7")
ax.axvline(-4.00, ls="--", lw=1, color="0.7")

ax.set_xlim(-4.15, -3.88)
ax.set_ylim(0.5, len(data) + 0.8)

ax.set_xlabel(r"$M_{\mathrm{TRGB}}$ [mag]", fontsize=12)
ax.set_title("TRGB zero-point calibrations from space telescopes", fontsize=13)

legend_elements = [
    Line2D([0], [0], marker="o", color="w",
           markerfacecolor="tab:blue", markersize=7, label="HST"),
    Line2D([0], [0], marker="o", color="w",
           markerfacecolor="tab:orange", markersize=7, label="JWST"),
]

ax.legend(handles=legend_elements, loc="lower right", frameon=False, fontsize=9)

plt.tight_layout()
plt.savefig("TRGB_space_telescopes_forest.png", dpi=300, bbox_inches="tight")
plt.show()
```
