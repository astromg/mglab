```python
# Statystyki obserwacji do raportu rocznego z OCM

import os
import numpy as np

from astropy.io import fits

path = "/work/vela/oca/fits/zb08/raw/"

nights = np.arange(677,1041,1)

#nights = np.arange(680,685,1)

files_number = 0
sci = 0
flat = 0
i = 0
j = 0
for n in nights:
    nht = f'{n:04d}'
    nht = path+nht
    if os.path.isdir(nht):
        i += 1
        j += 1
        fits_list = os.listdir(nht)
        f_cnt = 0
        flat_cnt = 0
        for f in fits_list:
            f_name = f'{nht}/{f}'
            hdul = fits.open(f_name)
            hdr = hdul[0].header
            hdul.close()
            files_number += 1
            if hdr["OBSTYPE"] == "science":
                f_cnt += 1
            if hdr["IMAGETYP"] == "flat":
                flat_cnt += 1
        if f_cnt > 200:
            sci += 1
        if flat_cnt > 10:
            flat += 1
        print(nht, f_cnt, flat_cnt)

    else:
        i += 1

print(i,j,sci,flat)
print("liczba plikow: ", files_number)
print("uptime: ", j/i)
print("science: ", sci/i)
print("flat: ", flat/i)


```
