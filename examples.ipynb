````python
# astropy FITS

from astropy.io import fits

f_name = "plik.fits"

hdul = fits.open(f_name)
data = hdul[0].data
hdr = hdul[0].header
hdul.close()
````

````python
# Plot FITS

import numpy as np
import matplotlib.pyplot as plt

image = data

vmin = np.median(image)  - 1 * np.std(image)
vmax = np.median(image)  + 1 * np.std(image)

plt.figure(figsize=(8, 8))
plt.imshow(image,vmin=vmin,vmax=vmax,cmap='binary')
plt.show()
````

````python
# tworzenie tabeli w fits

import numpy as np
from astropy.io import fits

# dane przykładowe
n = 5
cols = fits.ColDefs([
    fits.Column(name='ID',     format='K', array=np.arange(1, n+1)),
    fits.Column(name='RA',     format='D', array=[10.123, 10.234, 10.345, 10.456, 10.567]),
    fits.Column(name='DEC',    format='D', array=[-5.12, -5.10, -5.08, -5.05, -5.02]),
    fits.Column(name='Vmag',   format='E', array=[15.2, 16.1, 14.8, 15.9, 16.5]),
    fits.Column(name='Jmag',   format='E', array=[14.1, 14.9, 13.8, 14.7, 15.3]),
    fits.Column(name='e_Vmag', format='E', array=[0.03, 0.05, 0.02, 0.04, 0.06]),
    fits.Column(name='e_Jmag', format='E', array=[0.02, 0.03, 0.02, 0.03, 0.04]),
])

hdu = fits.BinTableHDU.from_columns(cols)
hdu.header['EXTNAME'] = 'CATALOG'
hdu.header['COMMENT'] = 'Example photometric catalog'

hdul = fits.HDUList([fits.PrimaryHDU(), hdu])
hdul.writeto('./data/example_catalog.fits', overwrite=True)
````

````python
# przyklady astropy odleglosci, wysokosci

from astropy.time import Time
import astropy.units as u
from astropy.coordinates import (SkyCoord, EarthLocation, AltAz, get_sun, get_body)

location = EarthLocation(lat=hdr["OBS-LAT"] * u.deg, lon=hdr["OBS-LONG"] * u.deg, height=hdr["OBS-ELEV"] * u.m)
obs_time = Time(hdr["DATE-OBS"], format="isot", scale="utc")
field = SkyCoord(ra=hdr["RA_TEL"] * u.deg, dec=hdr["DEC_TEL"] * u.deg, frame="icrs")

sun = get_sun(obs_time)
moon = get_body("moon", obs_time, location)

altaz_frame = AltAz(obstime=obs_time, location=location)
sun_altaz = sun.transform_to(altaz_frame)
moon_altaz = moon.transform_to(altaz_frame)

print(f"Wysokość Słońca: {sun_altaz.alt.deg:.2f} deg")
print(f"Wysokość Księżyca: {moon_altaz.alt.deg:.2f} deg")

sun_sep = field.separation(sun)
moon_sep = field.separation(moon)

print(f"Odległość pola od Słońca: {sun_sep.deg:.2f} deg")
print(f"Odległość pola od Księżyca: {moon_sep.deg:.2f} deg")



````
