````python
# lista plikow w katalogu

import os

file_list = []
for f in os.listdir("."):
    if f.endswith(".pdf"):
        file_list.append(f)

print(file_list)
````

````python
# lista plikow w katalogu i podkatalogach

import os

file_list = []

for root, dirs, files in os.walk("."):
    for f in files:
            if f.endswith(".psf"):
                file_list.append(root+"/"+f)

print(file_list)
````

````python
# Pasek postepu

import numpy
from tqdm import tqdm

data = numpy.arange(0,43,0.1)

res = 0
for i, val in enumerate(tqdm(data, desc="Processing")): 
    res = res + val

print(res)
    
````

````python
# Pasek postepu -  przyklad zaawansowany

import numpy
from tqdm import tqdm

bar = tqdm(total=100)

bar.set_description(f"Step {i}")
bar.update(0)
print("jeden")
bar.update(20)
print("dwa")
bar.update(40)
print("finish")

bar.close()
````

````python
# astropy FITS

from astropy.io import fits

f_name = "plik.fits"

hdul = fits.open(f_name)
data = hdul[0].data
hdr = hdul[0].header
hdul.close()
````

````python
# Plot FITS

import numpy as np
import matplotlib.pyplot as plt

image = data

vmin = np.median(image)  - 1 * np.std(image)
vmax = np.median(image)  + 1 * np.std(image)

plt.figure(figsize=(8, 8))
plt.imshow(image,vmin=vmin,vmax=vmax,cmap='binary')
plt.show()
````
